```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/all_video_games(cleaned).csv")
df
```

```python
# Menghapus nilai NaN
df = df.fillna('')
df
```

```python
df.describe() #menghasilkan ringkasan statistik deskriptif dari data numerik
```

```python
df.info() #memberikan gambaran umum tentang struktur data
```

```python
df.isnull().sum() #menghitung jumlah nilai yang hilang (null) dalam setiap kolom
```

```python
df = df.drop("Platforms Info", axis = 1) # menghilangkan kolom
df
```

```python
df["Product Rating"] = df["Product Rating"].str.replace("Rated E For Everyone", "Rated E")
df.head()
```

```python
plt.figure(figsize= (17,6)) #mengatur ukuran gambar
ax = sns.countplot(data = df, x = "Product Rating") #membuat diagram batang
ax.bar_label(ax.containers[0]) #menambahkan label di atas setiap batang pada countplot dengan nilai frekuensi masing-masing kategori
plt.title("Product Rating Distribution") #membuat judul diagram
plt.show()
```

```python
# Mengkonversi kolom 'User Score' dan 'User Ratings Count' ke tipe data numerik
df['User Score'] = pd.to_numeric(df['User Score'], errors='coerce')
df['User Ratings Count'] = pd.to_numeric(df['User Ratings Count'], errors='coerce')

# Menghapus baris yang memiliki nilai NaN di kedua kolom tersebut
df.dropna(subset=['User Score', 'User Ratings Count'], inplace=True)

# Mengelompokkan baris DataFrame berdasarkan nilai dalam kolom "Product Rating"
# Menerapkan fungsi agregasi ke kolom "User Score" dan "User Ratings Count"
gb = df.groupby("Product Rating").agg({"User Score": 'mean', "User Ratings Count": "mean"})
print(gb)
```

```python
plt.figure(figsize= (6,6))
sns.heatmap(gb, annot = True) #parameter annot=True digunakan untuk menampilkan nilai-nilai aktual di dalam sel heatmap
plt.title("Relationship between Product Rating and User")
plt.show
```

```python
gb1 = df.groupby("Genres").agg({"User Score":'mean',"User Ratings Count":"mean"})
print(gb1)
```

```python
plt.figure(figsize= (6,6))
sns.heatmap(gb1.head(10), annot = True)
plt.title("Relationship between Genres and User")
plt.show
```

```python
sns.boxplot(data = df, x = "User Score") #membuat box plot yang menampilkan distribusi dari nilai-nilai dalam kolom "User Score" dari DataFrame
plt.show
```

```python
plt.figure(figsize= (16,5))
sns.boxplot(data = df, x = "Product Rating")
plt.show
```

```python
print(df["Product Rating"].unique()) #mencetak semua nilai unik yang ada dalam kolom "Product Rating"
['Rated E' 'Rated T For Teen' 'Rated E +10 For Everyone +10' 'Rated M For Mature' 'Rated RP For Rate Pending' 'Rated AO For Adults Only']
```

```python
RatedE = df.loc[df['Product Rating'] == "Rated E"].count()
RatedT = df.loc[df['Product Rating'] == "Rated T For Teen"].count()
RatedE10 = df.loc[df['Product Rating'] == "Rated E +10 For Everyone +10"].count()  # Memperbaiki nama variabel
RatedM = df.loc[df['Product Rating'] == "Rated M For Mature"].count()
RatedRP = df.loc[df['Product Rating'] == "Rated RP For Rate Pending"].count()
RatedAO = df.loc[df['Product Rating'] == "Rated AO For Adults Only"].count()

print(RatedE["Product Rating"])

# List untuk label pada pie chart
l = ["Rated E", "Rated T For Teen", "Rated E +10 For Everyone +10", "Rated M For Mature", "Rated RP For Rate Pending", 'Rated AO For Adults Only']
# List untuk nilai pada pie chart
mlist = [RatedE["Product Rating"], RatedT["Product Rating"], RatedE10["Product Rating"], RatedM["Product Rating"], RatedRP["Product Rating"], RatedAO["Product Rating"]]

#visualisasi menggunakan fungsi plt.pie() dengan memasukkan list mlist sebagai nilai, dan list l sebagai label pada pie chart
print(mlist)
plt.figure(figsize= (10,10))
plt.pie(mlist, labels=l, autopct="%1.2f%%")
plt.title('Distribution of Product Rating')
plt.show()
```
