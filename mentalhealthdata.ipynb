```python
df = df.rename(columns = {
    'self_employed':'Self_Employed',
    'family_history':'Family_History',
    'treatment':'Treatment',
    'mental_health_interview':'Mental_Health_Interview',
    'care_options':'Care_Options',
})
```

```python
grouped_data = df.groupby(['Mental_Health_History',
       'Mental_Health_Interview'])
```

```python
aggregated_data = grouped_data.size().reset_index(name='Count')
print(aggregated_data)
```

```python
# Menggabungkan data dari kolom 'Days_Indoors' dan 'Growing_Stress'
data_combined = [df['Mental_Health_History'], df['Mental_Health_Interview']]

# Plot histogram
plt.figure(figsize=(5, 5))
plt.hist(data_combined, bins=20, color=['skyblue', 'salmon'], edgecolor='black', label=['Days Indoors', 'Growing Stress'])
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Mental_Health_History and Mental_Health_Interview')
plt.legend()
plt.show()
```

```python
# Kolom-kolom yang akan digunakan untuk diagram batang
columns = ['Gender', 'Country', 'Occupation', 'Self_Employed',
           'Family_History', 'Treatment', 'Days_Indoors', 'Growing_Stress',
           'Changes_Habits', 'Mental_Health_History', 'Mood_Swings',
           'Coping_Struggles', 'Work_Interest', 'Social_Weakness',
           'Mental_Health_Interview', 'Care_Options']

# Menentukan ukuran grid
num_cols = 3
num_rows = (len(columns) - 1) // num_cols + 1  # Jumlah baris dihitung berdasarkan jumlah kolom

# Ukuran plot
plt.figure(figsize=(16, 18))

# Membuat subplot untuk setiap kolom
for i, column in enumerate(columns, 1):
    plt.subplot(num_rows, num_cols, i)
    df[column].value_counts().plot(kind='bar', color='skyblue')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.title(f'Bar Plot of {column}')

# Menyesuaikan layout dan menampilkan plot
plt.tight_layout()
plt.show()
```

```python
# Kolom-kolom yang akan digunakan untuk boxplot
columns = ['Gender', 'Occupation', 'Self_Employed',
           'Family_History', 'Treatment', 'Days_Indoors', 'Growing_Stress',
           'Changes_Habits', 'Mental_Health_History', 'Mood_Swings',
           'Coping_Struggles', 'Work_Interest', 'Social_Weakness',
           'Mental_Health_Interview', 'Care_Options']

# Menentukan ukuran grid
num_cols = 4
num_rows = (len(columns) - 1) // num_cols + 1  # Jumlah baris dihitung berdasarkan jumlah kolom

# Ukuran plot
plt.figure(figsize=(16, 12))

# Membuat subplot untuk setiap kolom
for i, column in enumerate(columns, 1):
    plt.subplot(num_rows, num_cols, i)
    sns.boxplot(y=df[column], color='skyblue')
    plt.ylabel(column)
    plt.title(f'Boxplot of {column}')

# Menyesuaikan layout dan menampilkan plot
plt.tight_layout()
plt.show()
```

```python
# Kolom-kolom yang akan digunakan untuk diagram pie
columns = ['Gender', 'Occupation', 'Self_Employed',
           'Family_History', 'Treatment', 'Days_Indoors', 'Growing_Stress',
           'Changes_Habits', 'Mental_Health_History', 'Mood_Swings',
           'Coping_Struggles', 'Work_Interest', 'Social_Weakness',
           'Mental_Health_Interview', 'Care_Options']

# Menentukan ukuran grid
num_cols = 4
num_rows = (len(columns) - 1) // num_cols + 1  # Jumlah baris dihitung berdasarkan jumlah kolom

# Ukuran plot
plt.figure(figsize=(16, 12))

# Membuat subplot untuk setiap kolom
for i, column in enumerate(columns, 1):
    plt.subplot(num_rows, num_cols, i)
    counts = df[column].value_counts()
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Pie Chart of {column}')

# Menyesuaikan layout dan menampilkan plot
plt.tight_layout()
plt.show()
```

```python
# Memilih kolom numerik untuk diagram scatter
numeric_columns = ['Days_Indoors', 'Growing_Stress']

# Menghasilkan diagram scatter untuk setiap pasangan kolom numerik
plt.figure(figsize=(12, 8))
sns.set(style="ticks")

for i, col1 in enumerate(numeric_columns):
    for j, col2 in enumerate(numeric_columns):
        if i != j:
            plt.subplot(len(numeric_columns), len(numeric_columns), i * len(numeric_columns) + j + 1)
            sns.scatterplot(x=col1, y=col2, data=df)
            plt.xlabel(col1)
            plt.ylabel(col2)

plt.tight_layout()
plt.show()
```
